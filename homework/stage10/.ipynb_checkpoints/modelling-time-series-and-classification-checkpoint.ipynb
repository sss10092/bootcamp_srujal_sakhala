```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, TimeSeriesSplit
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
np.random.seed(42)
```

```python
dates = pd.date_range(start="2020-01-01", periods=500, freq="D")
returns = np.random.normal(0, 0.01, size=len(dates))
```

```python
df = pd.DataFrame({"ret": returns}, index=dates)
```

```python
df["lag1"] = df["ret"].shift(1)
```

```python
df["roll_mean_5"] = df["ret"].rolling(5).mean()
df["roll_std_5"] = df["ret"].rolling(5).std()
```

```python
features = ["lag1", "roll_mean_5", "roll_std_5"]
```

```python
df["y_reg"] = df["ret"].shift(-1)
df_reg = df.dropna().copy()
```

```python
split_idx = int(len(df_reg) * 0.8)
train_reg, test_reg = df_reg.iloc[:split_idx], df_reg.iloc[split_idx:]
```

```python
X_train_reg, y_train_reg = train_reg[features], train_reg["y_reg"]
X_test_reg, y_test_reg = test_reg[features], test_reg["y_reg"]
```

```python
reg_pipeline = Pipeline([
("scaler", StandardScaler()),
("model", LinearRegression())
])
```

```python
reg_pipeline.fit(X_train_reg, y_train_reg)
y_pred_reg = reg_pipeline.predict(X_test_reg)
```

```python
mae = mean_absolute_error(y_test_reg, y_pred_reg)
rmse = np.sqrt(mean_squared_error(y_test_reg, y_pred_reg))
print("Regression MAE:", mae)
print("Regression RMSE:", rmse)
```

```python
plt.figure(figsize=(10,4))
plt.plot(y_test_reg.index, y_test_reg, label="Truth")
plt.plot(y_test_reg.index, y_pred_reg, label="Predicted")
plt.legend()
plt.title("Forecasting Next-Step Return")
plt.show()
```

```python
df["y_cls"] = (df["ret"].shift(-1) > 0).astype(int)
df_cls = df.dropna().copy()
```

```python
split_idx = int(len(df_cls) * 0.8)
train_cls, test_cls = df_cls.iloc[:split_idx], df_cls.iloc[split_idx:]
```

```python
X_train_cls, y_train_cls = train_cls[features], train_cls["y_cls"]
X_test_cls, y_test_cls = test_cls[features], test_cls["y_cls"]
```

```python
cls_pipeline = Pipeline([
("scaler", StandardScaler()),
("clf", LogisticRegression())
])
```

```python
cls_pipeline.fit(X_train_cls, y_train_cls)
y_pred_cls = cls_pipeline.predict(X_test_cls)
```

```python
acc = accuracy_score(y_test_cls, y_pred_cls)
prec = precision_score(y_test_cls, y_pred_cls)
rec = recall_score(y_test_cls, y_pred_cls)
f1 = f1_score(y_test_cls, y_pred_cls)
cm = confusion_matrix(y_test_cls, y_pred_cls)
```

```python
print("Classification Accuracy:", acc)
print("Precision:", prec)
print("Recall:", rec)
print("F1:", f1)
print("Confusion Matrix:\n", cm)
```

```python
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

 - The MAE (~0.0086) and RMSE (~0.0106) are small in absolute terms because returns are centered near zero.
 - Predictions roughly track the scale of returns, but due to noise, the model provides limited forecasting power.
 - This is expected: financial returns are notoriously hard to predict with simple linear models.


 **Classification Task:**
 - Accuracy (~45%) is worse than random guessing (50%), showing the logistic regression model struggles to identify directional patterns.
 - Precision (~0.44) and recall (~0.59) indicate the model catches some upward moves, but with many false positives.
 - The confusion matrix (18 TN, 35 FP, 19 FN, 27 TP) highlights a strong imbalance in misclassifications.


 **Assumptions and Challenges:**
 - Linear relationships may not hold in noisy financial data.
 - Features like lag and rolling stats provide limited signal; returns are close to white noise.
 - Non-stationarity further breaks assumptions, leading to poor out-of-sample generalization.


 **Next Steps:**
 - Enrich feature space with momentum, z-scores, or external variables.
 - Try nonlinear models (tree-based, ensemble, or recurrent neural networks).
 - Use walk-forward validation for more robust time-aware evaluation.
